```python
import pandas as pd
import json
from IPython.display import display, HTML

# Path file kamu
file_path = r'C:\Local D\Galeri Belajar\Project\SR_02\data_prep_for_sft\data\sft_dataset\sft_batch_001.jsonl'

def dataset_viewer(path, limit=10):
    records = []
    with open(path, 'r', encoding='utf-8') as f:
        for i, line in enumerate(f):
            if i >= limit: break
            if line.strip():
                data = json.loads(line)
                meta = data.get('metadata', {})
                msgs = data.get('messages', [])
                
                # Ekstrak konten percakapan
                user_msg = next((m['content'] for m in msgs if m['role'] == 'user'), "")
                assistant_msg = next((m['content'] for m in msgs if m['role'] == 'assistant'), "")
                
                records.append({
                    "ID": meta.get('source_chunk_id'),
                    "Mapel": meta.get('mapel'),
                    "Bab": meta.get('bab'),
                    "Sub-bab": meta.get('sub_bab'),
                    "Input (Siswa)": user_msg,
                    "Output (Ahli Konten)": assistant_msg,
                    "Model": meta.get('model_used')
                })
    
    df = pd.DataFrame(records)
    
    # Render ke HTML dengan CSS untuk scrollable box
    html_style = """
    <style>
        .renderjson { font-family: sans-serif; }
        th { background-color: #f2f2f2; position: sticky; top: 0; }
        td { max-width: 400px; max-height: 150px; overflow-y: auto; text-align: left !important; vertical-align: top !important; }
        .content-box { max-height: 200px; overflow-y: auto; white-space: pre-wrap; font-size: 13px; line-height: 1.5; }
    </style>
    """
    
    # Gunakan formatter untuk membuat konten di dalam sel bisa di-scroll
    formatters = {
        'Input (Siswa)': lambda x: f'<div class="content-box">{x}</div>',
        'Output (Ahli Konten)': lambda x: f'<div class="content-box">{x}</div>'
    }
    
    display(HTML(html_style + df.to_html(escape=False, formatters=formatters, index=False)))

# Jalankan viewer
dataset_viewer(file_path, limit=5)
```
